#### Q3. Explain the use of the following operators using an example:
    (i) /
    (ii) %
    (iii) //
    (iv) **

#### Ans.
Above mentioned operators are frequently used operators. There explanation is given as below: 

    (i) '/' - Division Operator:
                            The division operator is used to divide one number by another.
                        Example: 
                            result = 10 / 2
                            print(result)
                        Output: 
                            5.0
    (ii) '%' - Modulo / Modulus Operator:
                            The modulus operator returns the remainder of the division of one number by another.
                        Example: 
                            remainder = 10 % 3
                            print(remainder)
                        Output:
                            1
    (iii) '//' - Floor Division Operator:
                        The floor division operator performs division and rounds down to the nearest integer.
                    Example: 
                        result = 10 // 3
                        print(result)

                    Output:
                        3

    (iv) '**' -  Exponentiation Operator:
                        The exponentiation operator raises a number to the power of another.

                    Example: 
                        result = 2 ** 3
                        print(result)
                    Output: 
                        8

#### Q7. What do you understand about mutable and immutable data types? Give examples for both showing this property.

#### Ans: 
    Mutable and immutable are terms used to describe the ability or inability to modify an object once it has been created. In the context of programming and data types, this distinction is essential.

   *  Mutable Data Types:

              Definition: Mutable data types are those whose values or contents can be changed or modified after the object is created.
          Examples:
             List: Lists in Python are mutable. You can modify, add, or remove elements after creating a list.
                     my_list = [1, 2, 3]
                     my_list[0] = 99
                     print(my_list)

                  Output: [99, 2, 3]
             Dictionary: Dictionaries are mutable as well. You can add, remove, or modify key-value pairs.
                  my_dict = {'a': 1, 'b': 2, 'c': 3}
                  my_dict['a'] = 99
                  print(my_dict)

                  Output: {'a': 99, 'b': 2, 'c': 3}
    * Immutable Data Types:
        Definition: Immutable data types, on the other hand, cannot be changed once they are created. Any operation that appears to modify an immutable object actually creates a new object.
            Examples:
                Tuple: Tuples are immutable in Python. Once a tuple is created, you cannot change its elements.
                
                    my_tuple = (1, 2, 3)
                    # The following line would result in an error:
                    # my_tuple[0] = 99
                
                String: Strings are immutable as well. You cannot modify the characters of a string after it has been created.
                    my_string = "hello"
                    # The following line would result in an error:
                    # my_string[0] = 'H'
                Integer, Float, and Other Numeric Types: Numbers are also immutable in Python.
                    my_number = 42
                    # The following line would result in an error:
                    # my_number = my_number + 1
    
    Understanding the mutability or immutability of data types is important a good and valid code, especially when dealing with functions or methods that involve modifying or returning modified versions of objects.
